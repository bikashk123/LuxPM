5. From the center of the image, in all directions, increase the RGB values of the pixels in a manner that, each pixel from the center, the percentage drops by 1%. i.e. the center pixel's RGB will increase by 50%, and the next pixels in x and y directions will be 49%. This goes on and on until the increase becomes 0 %.

```python
import cv2
import numpy as np
import imutils
import matplotlib.pyplot as plt
from PIL import Image 
```

## Load image

```python
img=cv2.imread('images.jpg')
```

```python
cv2.imshow('image', img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 1.Transform the image in the +x direction by 25%, and create an image

```python
rows, cols, dim = img.shape
print(rows,cols,dim)
```

```python
# transformation matrix for translation
M = np.float32([[1, 0.25, 0],
                [0, 1  , 0],
                [0, 0  , 1]])
```

```python
translated_img = cv2.warpPerspective(img, M, (cols, rows))
plt.axis('off') # disable x & y axis
plt.imshow(translated_img) # show the resulting image
plt.show()
# save the resulting image 
plt.imsave("translated_x_25%.jpg", translated_img)
```

### 2.Transform the image in the +y direction by 25%, and create an image

```python
# shearing applied to y-axis
N = np.float32([[1,   0, 0],
              [0.25, 1, 0],
              [0,   0, 1]])
```

```python
translated_img = cv2.warpPerspective(img, N, (cols, rows))
# disable x & y axis
plt.axis('off')
plt.imshow(translated_img) # show the resulting image
plt.show()
# save the resulting image 
plt.imsave("translated_y_25%.jpg", translated_img)
```

### 3. Rotate the image in Z by 90 degree

```python
image = cv2.imread(r"images.jpg")
rot = imutils.rotate(image, angle=90)
cv2.imshow("Rotated", rot)
cv2.imwrite("Rotated_z_90.jpg",rot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#another approach

img = Image.open(r"images.jpg") 
rotate_img= img.rotate(90)
im1 = rotate_img.save("rotate_by_90.jpg")
rotate_img.show() 
```

### 4. Rotate the image in Z by -90 degree

```python
image = cv2.imread(r"images.jpg")
rot = imutils.rotate(image, angle=-90)
cv2.imshow("Rotated", rot)
cv2.imwrite("Rotated_z_-90.jpg",rot)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#another approach
img = Image.open(r"images.jpg") 
rotate_img= img.rotate(-90)
im1 = rotate_img.save("rotate_by_-90.jpg")
rotate_img.show() 
```
